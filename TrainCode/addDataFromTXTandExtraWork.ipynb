```python
def eliminar_archivos_con_copia(path,nombres_ficheros):
    for nombre_fichero in nombres_ficheros:
        if not "_da" in nombre_fichero:
            try:
                os.remove(path+"/"+nombre_fichero)
            except OSError as e:
                print(f"No se pudo eliminar el archivo {nombre_fichero}: {e}")
```

```python
import os

carpeta = 'DAGPT_isolated/ironic'  # Cambia 'ruta/a/la/carpeta' por la ruta de tu carpeta

for nombre_archivo in os.listdir(carpeta):
    # Obtener la ruta completa del archivo
    ruta_completa = os.path.join(carpeta, nombre_archivo)

    # Comprobar si es un archivo
    if os.path.isfile(ruta_completa):
        # Dividir el nombre del archivo y la extensión
        nombre, extension = os.path.splitext(nombre_archivo)

        # Crear el nuevo nombre con la palabra "_da"
        nuevo_nombre = nombre + '_da' + extension

        # Obtener la ruta completa del nuevo nombre de archivo
        nueva_ruta_completa = os.path.join(carpeta, nuevo_nombre)

        # Renombrar el archivo
        os.rename(ruta_completa, nueva_ruta_completa)
```

```python
import re

def reemplazar_hashtags_usuarios_urls(ruta_archivo):
    with open(ruta_archivo, 'r+') as archivo:
        contenido = archivo.read()

        # Reemplazar hashtags por 'hashtag'
        contenido = re.sub(r'#\w+', 'hashtag', contenido)

        # Reemplazar nombres de usuario por 'usuario'
        contenido = re.sub(r'@\w+', 'usuario', contenido)

        # Reemplazar URLs por 'URL'
        contenido = re.sub(r'http\S+|www\S+', 'URL', contenido)

        # Regresar al inicio del archivo para sobrescribir el contenido
        archivo.seek(0)

        # Sobrescribir el contenido modificado
        archivo.write(contenido)

        # Truncar el contenido adicional si es necesario
        archivo.truncate()

        print(f'Se ha modificado el archivo {nombre_archivo}.')

```
