```python
from googletrans import Translator

def translate(initial_sentence):
    translator = Translator()
    # traduccion = translator.translate(initial_sentence, dest="en")
    # traduccion = translator.translate(initial_sentence, dest="de")
    traduccion = translator.translate(initial_sentence, dest="ko")
    #traduccion = translator.translate(traduccion.text, dest="ru")
    # traduccion = translator.translate(traduccion.text, dest="pt")
    traduccion = translator.translate(traduccion.text, dest="es")

    return traduccion.text


def duplicate_data(folder_path,folder_path_dest):
    for filename in os.listdir(folder_path):
        if os.path.isfile(os.path.join(folder_path, filename)):

            new_filename = os.path.splitext(filename)[0] + 'copia' + os.path.splitext(filename)[1]

            shutil.copy(os.path.join(folder_path, filename), os.path.join(folder_path, new_filename))

            with open(os.path.join(folder_path, filename), 'r') as original_file:
                content = original_file.read()

            content = translate(content)
            with open(os.path.join(folder_path_dest, new_filename), 'w') as new_file:
                new_file.write(content)
```

```python
import os

def eliminar_archivos_con_copia(path,nombres_ficheros):
    for nombre_fichero in nombres_ficheros:
        if "copia" in nombre_fichero:
            try:
                os.remove(path+"/"+nombre_fichero)
            except OSError as e:
                print(f"No se pudo eliminar el archivo {nombre_fichero}: {e}")

# Ejemplo de uso
# nombres_ficheros = ["archivo1.txt", "archivo2.copia.txt", "archivo3.docx", "archivo4.copia.docx"]

# auxborrarcopias1 =  os.listdir("datasetFINAL/train_ds/ironic")
# eliminar_archivos_con_copia("datasetFINAL/train_ds/ironic",auxborrarcopias1)

# auxborrarcopias2 =  os.listdir("datasetFINAL/train_ds/notironic")
# eliminar_archivos_con_copia("datasetFINAL/train_ds/notironic",auxborrarcopias2)
```

```python
def generar_archivos_frases(archivo_entrada,id):
  # Abrir el archivo de entrada en modo lectura
  with open(archivo_entrada, 'r') as archivo:
      contenido = archivo.read()

      # Dividir el contenido en frases utilizando saltos de línea como separadores
      frases = contenido.split('\n')

      # Eliminar las frases vacías
      frases = [frase.strip() for frase in frases if frase.strip()]

      # Generar un archivo de salida para cada frase
      for i, frase in enumerate(frases):
          nombre_archivo_salida = f'dab{id}_{i + 1}.txt'

          # Escribir la frase en el archivo de salida
          with open("DAisoGPsupervized/BATCH2/"+nombre_archivo_salida, 'w') as archivo_salida:
              archivo_salida.write(frase)
          print(nombre_archivo_salida)
```

```python
import os
FILES =  os.listdir("DAisoGPsupervized")
a = 0
for filename in FILES:
  if(".txt" in filename):
    generar_archivos_frases("DAisoGPsupervized/"+filename, a)
    a += 1
```
