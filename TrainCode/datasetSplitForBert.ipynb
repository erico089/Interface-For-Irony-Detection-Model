```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split

on = True

#augmentation_info = ""
augmentation_info = pd.DataFrame(columns=['SENTENCE', 'TYPE','IRONIC'])
# augmentation_info_counter = 0

from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/MyDrive/TFG - Irony Detection - Èric Casanovas Pérez/Code/Tests Bert
df_test = pd.DataFrame()

!ls

if on:
  #Primero usar df i hacer las particiones correspondientes
  #Crear las carpetas
  !mkdir datasetFINAL
  !mkdir datasetFINAL/train_ds
  !mkdir datasetFINAL/train_ds/ironic
  !mkdir datasetFINL/train_ds/notironic
  !mkdir datasetFINALA/test_ds
  !mkdir datasetFINAL/test_ds/ironic
  !mkdir datasetFINAL/test_ds/notironic
  !mkdir datasetFINAL/validation_ds
  !mkdir datasetFINAL/validation_ds/ironic
  !mkdir datasetFINAL/validation_ds/notironic

  !mkdir dataseDA1_real_test

  df = pd.read_csv('ds.csv')
  df = df.drop(['ID'], axis=1)

  df = df.sample(frac=1).reset_index(drop=True)

  dftrain, dftest = train_test_split(df, test_size=0.2)

  # First lets do the partitions for the not ironic dataset
  df_not_ironic = dftrain[df.IS_IRONIC == 0]
  df_not_ironic = df_not_ironic.reset_index()
  not_iro_len = df_not_ironic[df_not_ironic.columns[0]].count()
  print(not_iro_len)

  for ind in df_not_ironic.index:
    if(ind <int(not_iro_len*0.8)):
      # augmentation_info.loc[len(augmentation_info)] = [df_not_ironic.iloc[ind]['MESSAGE'], df_not_ironic.iloc[ind]['TOPIC'],'NO IRONICO']
      # augmentation_info = augmentation_info+"\n Frase: "+df_not_ironic.iloc[ind]['MESSAGE']+" Topic : "+df_not_ironic.iloc[ind]['TOPIC'] + "\n"
      with open('datasetDA1/train_ds/notironic/'+str(ind) +'.txt', 'w') as writefile:
        writefile.write(df_not_ironic.iloc[ind]['MESSAGE'])

    elif(ind < (int(not_iro_len*0.8)+int(not_iro_len*0.2*0.8)) ):
      df_test = df_test.append({'IS_IRONIC':df.iloc[ind]['IS_IRONIC'], 'MESSAGE':df.iloc[ind]['MESSAGE']}, ignore_index=True)
      with open('datasetDA1/test_ds/notironic/'+str(ind) +'.txt', 'w') as writefile:
        writefile.write(df_not_ironic.iloc[ind]['MESSAGE'])

    else:
      with open('datasetDA1/validation_ds/notironic/'+str(ind) +'.txt', 'w') as writefile:
        writefile.write(df_not_ironic.iloc[ind]['MESSAGE'])

  # The same for the ironic dataset
  df_ironic = dftrain[df.IS_IRONIC == 1]
  df_ironic = df_ironic.reset_index()
  iro_len = df_ironic[df_ironic.columns[0]].count()
  print(iro_len)

  for ind in df_ironic.index:
    if(ind <int(iro_len*0.8)):
      # augmentation_info.loc[len(augmentation_info)] = [df_ironic.iloc[ind]['MESSAGE'], df_ironic.iloc[ind]['TOPIC'],'IRONICO']
      # augmentation_info = augmentation_info+"\n Frase: "+df_not_ironic.iloc[ind]['MESSAGE']+" Topic : "+df_not_ironic.iloc[ind]['TOPIC'] + "\n"
      with open('datasetDA1/train_ds/ironic/'+str(ind) +'.txt', 'w') as writefile:
        writefile.write(df_ironic.iloc[ind]['MESSAGE'])

    elif(ind < (int(iro_len*0.8)+int(iro_len*0.2*0.8)) ):
      df_test = df_test.append({'IS_IRONIC':df.iloc[ind]['IS_IRONIC'], 'MESSAGE':df.iloc[ind]['MESSAGE']}, ignore_index=True)
      with open('datasetDA1/test_ds/ironic/'+str(ind) +'.txt', 'w') as writefile:
        writefile.write(df_ironic.iloc[ind]['MESSAGE'])

    else:
      with open('datasetDA1/validation_ds/ironic/'+str(ind) +'.txt', 'w') as writefile:
        writefile.write(df_ironic.iloc[ind]['MESSAGE'])


  #For the test
  test_len = dftest[dftest.columns[0]].count()

  dftest = dftest.reset_index(drop=True)
  for ind in dftest.index:
     with open('dataseDA1_real_test/'+str(ind) + str(dftest.iloc[ind]['IS_IRONIC']) +'.txt', 'w') as writefile:
        writefile.write(dftest.iloc[ind]['MESSAGE'])

  print(test_len)

  # de cada linea hacer un txt con la linea
```
